```python
#while loop

#it will execute a block of code untill a condition is true

count = 1

while count < 10: #condition #iter1 = 1, iter2 = 3, iter3 = 5, 7
    print(count) 
    count += 2 #count 3, count = 3 + 2 = 5, count = 5 + 2 = 7
```

```python
count = 1

while count < 10: #condition #iter1 = 1, iter2 = 3, iter3 = 5, 7
    if count == 3:
        break
    print(count)
    count = count + 2 
```

```python
count = 1
while count < 10: #condition #iter1 = 1, iter2 = 3, iter3 = 5, 7
    count += 2
    if count < 3:
        continue
    print(count) 
```
