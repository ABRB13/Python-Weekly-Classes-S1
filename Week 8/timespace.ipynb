```python
import time
def sum_of_number(n):
    total_sum = 0

   

    for i in range(1, 1+n):
        total_sum += i
        
    return total_sum
    


```

```python
sum_of_number(10000000)
```

```python
import matplotlib.pyplot as plt 
import time 

def tns_complex():
    input_sizes = []
    time_complexity = []
    space_complexity = [] 

    for n in range(100, 10000, 2000):
        start_time = time.time()

        result = sum_of_number(n)
        
        end_time = time.time()

        total_time = end_time - start_time 

        memory_usage = n * 8 
        input_sizes.append(n)
        time_complexity.append(total_time)
        space_complexity.append(memory_usage)

    plt.figure(figsize = (10,10))
    plt.plot(input_sizes , time_complexity, label = "time complex")

    plt.xlabel('input_sizes(n)')
    plt.ylabel('time')
    plt.title('time')

    plt.figure(figsize = (10,10))
    plt.plot(input_sizes , space_complexity, label = "time complex")

    plt.xlabel('input_sizes(n)')
    plt.ylabel('space')
    plt.title('space')

    plt.legend()
    plt.show()


```

```python
tns_complex()
```

```python
def binarySearch(arr, l, r, x):

    while l <= r:

        mid = l + (r - l) // 2

        # Check if x is present at mid
        if arr[mid] == x:
            return mid

        # If x is greater, ignore left half
        elif arr[mid] < x: 
            l = mid + 1

        # If x is smaller, ignore right half
        else:
            r = mid - 1
```

```python
arr = [2, 3, 4, 10, 40]
x = 40

result = binarySearch(arr, 0, len(arr)-1, x)

```

```python
#binary search 

import matplotlib.pyplot as plt 
import time 

def tns_complex():
    array_sizes = [1,2,3,4,5,6,7]
    input_sizes = []
    time_complexity = []
    space_complexity = [] 

    for n in range(100, 10000, 2000):

        array_sizes.append(n)


        start_time = time.time()

        result = binarySearch(array_sizes, 0, len(array_sizes)-1, n)
        
        end_time = time.time()

        total_time = end_time - start_time 

        memory_usage = n * 8 
        
        input_sizes.append(len(array_sizes))
        time_complexity.append(total_time)
        space_complexity.append(memory_usage)

    plt.figure(figsize = (10,10))
    plt.plot(input_sizes , time_complexity, label = "time complex")

    plt.xlabel('input_sizes(n)')
    plt.ylabel('time')
    plt.title('time')

    plt.figure(figsize = (10,10))
    plt.plot(input_sizes , space_complexity, label = "time complex")

    plt.xlabel('input_sizes(n)')
    plt.ylabel('space')
    plt.title('space')

    plt.legend()
    plt.show()


```
